This notebook will show you how to create and query a table or DataFrame that you uploaded to DBFS. [DBFS](https://docs.databricks.com/user-guide/dbfs-databricks-file-system.html) is a Databricks File System that allows you to store data for querying inside of Databricks. This notebook assumes that you have a file already inside of DBFS that you would like to read from.

This notebook is written in **Python** so the default cell type is Python. However, you can use different languages by using the `%LANGUAGE` syntax. Python, Scala, SQL, and R are all supported.

```
import pyspark
from pyspark.sql import SparkSession
```

```
# File location and type
file_location = "/FileStore/tables/heart-3.csv"
file_type = "csv"

# CSV options
infer_schema = "false"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df)
```

```
# Create a view or table

temp_table_name = "tblheart"

df.createOrReplaceTempView(temp_table_name)
```

```
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.stat import Correlation


inputCols = ["Age", "ChestPainType",  "RestingBloodPressure","Cholestoral" , "FastingBloodSugar", "MaxHeartRateAchived", "ExerciseInducedAngina", "STDepress", "slope", "NumMajorVessel", "ThaliumStressTest", "PredictedHeartDisease"]
vector_col = "corr_features"

assembler = VectorAssembler(inputCols = inputCols, outputCol = vector_col)
myGraph_vector = assembler.transform(dfHeartNum).select(vector_col)
matrix = Correlation.corr(myGraph_vector, vector_col).collect()[0][0]
corrmatrix = matrix.toArray().tolist()

def plot_corr_matrix(correlations,attr):
     print(attr) 
     sns.set_context('paper')
     plt.figure(figsize=(25,20)) 
     ax=sns.heatmap(corrmatrix,  cmap="bwr", annot=True, linewidths=.5, xticklabels = attr, yticklabels = attr,
                      cbar_kws={"label": "Pearson correlation coefficient "}, vmin=-1, vmax=1)
     ax.set_title("Correlation Matrix for Specified Attributes")

plot_corr_matrix(corrmatrix, inputCols)
```

### ML testing with Random Forest Classifier

```
display(dfHeartNum)
```

```
#defining Features
nonFeatureCols =["PredictedHeartDisease"]
features = [a for a in dfHeartNum.columns if a not in nonFeatureCols]
```

```
#Vector Assembler
from pyspark.ml.feature import VectorAssembler

assembler = (VectorAssembler()
  .setInputCols(features)
  .setOutputCol("features"))

finalPrep = assembler.transform(dfHeartNum)
finalPrep.show()
```

```
#train test split
train,test = finalPrep.randomSplit([0.7,0.3], seed = 3)

train.cache()
test.cache()

print(train.count()) 
print(test.count())
```

```
#import modules and RF Classifier
from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator
from pyspark.ml import Pipeline
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

rfclassifier = (RandomForestClassifier(labelCol="PredictedHeartDisease",
                                       featuresCol="features", numTrees = 10)
               )
```

```
#fit model
model = rfclassifier.fit(train)
```

```
#predict the test set
predictions = model.transform(test)
```

```
#check predictions
predictions.select("PredictedHeartDisease",'rawPrediction', 'prediction', 'probability').show()
```

```
#evaluate accuracy
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

evaluator = MulticlassClassificationEvaluator(labelCol="PredictedHeartDisease", predictionCol="prediction")
accuracy = evaluator.evaluate(predictions)
print("Accuracy = %s" % (accuracy))
print("Test Error = %s" % (1.0 - accuracy))
```

## Trying a param gridbuilder and Cross Validator on RF

```
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator

rfclassifier_grid_instance = RandomForestClassifier(
  labelCol="PredictedHeartDisease",
  featuresCol="features"
  )
               
paramgrid_variables = (ParamGridBuilder()
                       .addGrid(rfclassifier_grid_instance.maxDepth, [5,10,15,20])
                       .addGrid(rfclassifier_grid_instance.numTrees, [20,10,30,5])
                       .build()
                      )

cv = CrossValidator(
  estimator = rfclassifier_grid_instance,
  estimatorParamMaps = paramgrid_variables,
  evaluator=evaluator,
  numFolds=2
)
                   

cv_model_1 = cv.fit(train)
```

```
cv_model_1.avgMetrics
```

```
cv_model_bestmodel = cv_model_1.bestModel
```

```
cv_predictions = cv_model_bestmodel.transform(test)
cv_predictions.select("PredictedHeartDisease",'rawPrediction','features', 'prediction', 'probability',).show()
```
